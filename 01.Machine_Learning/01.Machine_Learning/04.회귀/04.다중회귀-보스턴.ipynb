# 다중 회귀 - 보스톤 주택 가격

```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['PRICE'] = boston.target
df.head()
```

- 산점도와 선형회귀선

```python
sns.regplot(x='CRIM', y='PRICE', data=df)
plt.show()
```

```python
fig, axs = plt.subplots(nrows=3, ncols=4, figsize=(16,12))
for i, feature in enumerate(df.columns[1:-1]):
    row, col = i//4, i%4
    sns.regplot(x=feature, y='PRICE', data=df, ax=axs[row][col])
```

- feature 별 R squared value

```python
from sklearn.linear_model import LinearRegression
```

```python
y = boston.target
X = df.CRIM.values.reshape(-1,1)
# X = df['CRIM'].values.reshape(-1,1)
lr = LinearRegression()
lr.fit(X, y)
lr.score(X, y)
```

```python
y = boston.target
for feature in boston.feature_names:
    X = df[feature].values.reshape(-1,1)
    lr = LinearRegression()
    lr.fit(X, y)
    r2 = lr.score(X, y)
    print(f'{feature}:\t{r2:.4f}')
```

- 다중 선형회귀

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    boston.data, boston.target, test_size=0.1, random_state=2021
)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 다중 선형회귀에서는 R squared value가 무의미
lr.score(X_test, y_test)
```

```python
# Weight
lr.coef_
```

```python
# bias
lr.intercept_
```

- 테스트 데이터

```python
X_test[0], y_test[0]
```

```python
# 기울기와 절편을 이용한 예측값 (Weight, Bias)
np.dot(X_test[0], lr.coef_) + lr.intercept_
```

```python
# predict로 예측한 값
lr.predict(X_test[:1]) # 슬라이싱을 해서 2차원 데이터로 출력됨 ~ X_test[0]
# lr.predict(X_test[i].reshape(1, -1))
# lr.predict([X_test[0]]) 
```

```python
# 실제값
y_test[0]
```

```python
# 5개 데이터에 대한 예측값 및 실제값
for i in range(10):
    pred1 = np.dot(X_test[i], lr.coef_) + lr.intercept_
    pred2 = lr.predict(X_test[i].reshape(1, -1))
    print(f'실제값: {y_test[i]}, 예측값1: {pred1:.2f}, 예측값2: {pred2[0]:.2f}')
```
