```python
s = pd.Series(['가나다 123', 'ㅠㅠ!!!', '한글alphabet'])
s
```

```python
# 정규 표현식(Regular expression)
import re
re.sub('[0123456789]', '.', '가나다 123')  # (정규 표현식, 뭘로 바꿀지, 바꿀 문자 )
# re.sub('[0-9]', '.', '가나다 123') 위와 같은 표현
```

```python
re.sub('[a-z]', '.', '한글alphabet')
```

```python
# 문자열 메소드인 replace는 정규표현식을 지원하지 않음
'한글alphabet'.replace('[a-z]','.')
```

```python
# 한글 정규 표현식
re.sub('[ㄱ-ㅎㅏ-ㅣ가-힣]', '.', '한글alphabet')
```

```python
# Pandas str.replace() 메소드는 정규 표현식을 지원함
s.str.replace('[ㄱ-ㅎㅏ-ㅣ가-힣]', '.')
```

```python
# ^ 는 not(부정)을 의미
s.str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣 ]', '')  # 한글과 공백을 제외한 글자는 삭제
```

- str.메소드로 천단위 구분기호를 숫자로 바꾸기

```python
cctv = pd.read_csv('Seoul_CCTV.csv', skiprows = 1, encoding='euc-kr')

cctv.head()
```

```python
for column in cctv.columns[1:]:
    cctv[column] = cctv[column].str.replace(',', '').astype(int)

cctv.head()
```
