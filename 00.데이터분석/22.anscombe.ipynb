```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False  # 그래프에서 한글 출력을 위한 코드
plt.rcParams['font.family'] = 'Malgun Gothic'  # 그래프에서 한글 출력을 위한 코드
```

```python
ans = sns.load_dataset('anscombe')
ans.head(), ans.tail()
```

```python
ans.shape
```

- dataset I 에 대해 그래프를 그리기

```python
ds = ans[ans.dataset == 'I'] 
ds
```

- 회귀식 y = 0.5 * x + 3

```python
plt.scatter(ds.x, ds.y)
plt.plot([4,14], [5, 10], 'r-', lw=2)
plt.title('Dataset I')
plt.show()
```

- 4개의 그래프를 하나의 그림에 그리기

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,8), sharex=True, sharey=True) 
```

```python
# i = 0
# for roman in ['I','II','III','IV']:
#     ax = axs[i//2, i%2] 
#     i += 1
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,8)) 
for i, roman in enumerate(['I','II','III','IV']): 
    ax = axes[i//2, i%2] 
    ds = ans[ans.dataset == roman] 
    ax.scatter(ds.x, ds.y) 
    ax.plot([3,20], [4.5, 13], 'r-', lw=2) # y = 0.5 * x + 3 
    ax.set_ylim(2,13)
    ax.set_title(f'Dataset {roman}') 
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(9,7), sharex=True, sharey=False) 
for i, roman in enumerate(['I','II','III','IV']): 
    ax = axes[i//2, i%2] 
    ds = ans[ans.dataset == roman] 
    ax.scatter(ds.x, ds.y, s=10) 
    ax.plot([3,20], [4.5, 13], 'r-', lw=2) # y = 0.5 * x + 3 
    # ax.set_ylim(2,13)
    ax.set_title(f'Dataset {roman}')  
fig.suptitle("Anscombe's Quartet", fontsize=10) 
plt.tight_layout() 
plt.show() 
```
